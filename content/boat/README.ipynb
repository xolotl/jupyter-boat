# Jupyter BOAT
Jupyter BOAT is a republication of [BOAT](https://xolotl.org/boat-1-0/), the bulk open attribution tool that enables you to generate open licensing/attribution statements from a list of basic information about a collection of works, but here using a Jupyter computational notebook environment instead of BOAT's original spreadsheet. For more information about the origin and development of BOAT, please see [all my BOAT posts](https://xolotl.org/tags/boat/).

**If you just want to see BOAT in action right away using example data, visit <a href="https://xolotl.github.io/jupyter-boat/lab/index.html?path=boat%2Fboat.ipynb" target="_self">`boat.ipynb`</a> in this notebook, choose `Run All Cells` from the notebook's `Run` menu, and scroll down past the code to see BOAT's output.**

Otherwise, read on to learn how Jupyter BOAT works and use it with data from your own collection of works.

## Using Jupyter BOAT
All BOAT files including this `README` are located in the `boat` folder in this notebook. You can modify and work with BOAT's files within the [the notebook I published](https://xolotl.org/jupyter-boat/) â€” that's what computational notebooks are for! Or you can visit [the Jupyter BOAT GitHub reposistory](https://github.com/xolotl/jupyter-boat) and review, copy, duplicate, and/or contribute to the project.

Jupyter BOAT is delivered via a standard [Jupyter Lite demo](https://jupyterlite.github.io/demo) deployed as a static [GitHub Pages](https://pages.github.com/) website. The default materials from that demo are included in this notebook along with BOAT. If you know what you are doing, you can likely also use Jupyter BOAT either in an entirely different Jupyter (or other) computational notebook, or some other computing enviroment that supports python.

### BOAT's files
There are four files in the `boat` folder in this notebook that enable BOAT to work. You don't have to understand them all to use BOAT, but it might help to know what role each file plays.
1. **`works.csv`**: A comma-delimited list of works for which BOAT will generate licensing/attribution statements. You can modify or add to the works in this list. For example, you can put your own collection of works in this file to have BOAT generate licensing/attribution statements for them.
1. **`ipstatements.csv`**: A comma-delimited list of the various intellectual property (IP) statements used in Jupyter BOAT. You can modify or add to the the statements in this list. For example, maybe you'd like to change the wording of the public domain (PD) or all rights reserved (ARR) statements.
1. **`boat.ipynb`**: The Jupyter notebook python code that makes BOAT work, matching the list of works with the list of intellectual property statuses to generate full licensing/attribution statements for those works, delivered in various formats using a template when BOAT runs. You can change this python code to modify how BOAT works.
1. **`template.html`**: The [jinja template](https://palletsprojects.com/p/jinja/) that generates the outputs from the other files in Jupyter BOAT. You can modify this template to change the output generated by BOAT.

### See BOAT work
Jupyter BOAT is designed to function without any modifications to demonstrate how it works.

Once you see how BOAT works, it is expected that you would provide your own list of works for which you want BOAT to generate licensing/attribution statements. If your works include some with intellectual property statuses that are not already in BOAT's IP statements file, or if you need to make changes to the licensing/attribution statements BOAT generates, you might need to adjust its IP statements file (`ipstatements.csv`), the output template (`template.html`), or perhaps even the logic in the BOAT python code file (`boat.ipynb`).

To see BOAT generate output using the example works already listed in the works.csv file, just open <a href="https://xolotl.github.io/jupyter-boat/lab/index.html?path=boat%2Fboat.ipynb" target="_self">`boat.ipynb`</a> in this notebook, choose `Run All Cells` from the notebook's `Run` menu, and scroll down past the code to see BOAT's output.

### Use BOAT on your own collection of works
To use BOAT to generate licensing/attribution statements for your own collection of works, just drag the `works.csv` file from the `boat` folder to your computer, edit it using your preferred CSV editor (eg, Excel, Google Sheets), and then drag your revised `works.csv` file back in to the `boat` folder in this notebook and `run all` in the `boat.ipynb` file again.

### Change or add to BOAT's IP statements
To modify how BOAT generates the IP statements in its output, you can make changes to or add new statements to the `ipstatements.csv` file using the same process used in modifying the `works.csv` file above.

To work as expected, every row in the IP statements file should have a unique value in the `IP Status` column that can also be used in the `works.csv` file to link a work to the right IP statement. Each IP status is repeated for each format (eg, html, md, txt, txturl) with a value in the `IP Snippet` column that provides the correctly formatted output for works with that IP status (see existing entries for examples).

### Modify BOAT's output
To change BOAT's output, you can modify its `template.html` file using the same process used in modifying the `works.csv` file above, but using an HTML code editor. When modifying the template, it may be helpful to learn how [jinja templates](https://palletsprojects.com/p/jinja/) work.

### Modify BOAT's logic
To change how BOAT combines works and IP statuses to generate licensing/attribution statements using a template, you can modify its `boat.ipynb` file using the same process used in modifying the `works.csv` file above, but using a code editor. When modifying the python notebook file, it may be helpful to learn how [python](https://www.python.org/) works.
